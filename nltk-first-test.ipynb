The first step is to <em>import</em> in the code the parts of the <em>nltk</em> library that we are going to use.

```python
import nltk

from nltk.book import *

from nltk.corpus import gutenberg
```


Among the others, we imported the <em>gutenberg</em> part of the <em>nltk</em> library, which contains all the texts and corpora of the Gutenberg project. 

We are going to focus on the Hamlet by Shakespeare. In particular, the next instruction is storing in the variable <em>hamlet</em> all the words of the Hamlet, in the form of a list.

```python
hamlet = gutenberg.words("shakespeare-hamlet.txt")
```

From now on, we can access this list via the variable. For example, we can compute the length of this list, as in the following:

```python
all_words = len(hamlet)
print(f"Hamlet contains {all_words} words\n")
```

Note the <em>len()</em> <strong>function</strong>: you give a list as a parameter, and it returns the length of this list. Here, the returned value is stored in another variable, <em>all_words</em>.

Also note the <em>print()</em> function: you give a string as parameter, and it prints it in the output console (here, the part below the code box). The string is composed of a <em>static</em> part ("Hamlet contains ... words") and a <em>dynamic</em> one ("{all_words}"), which is replaced with the actual content of the variable. The initial "f" and the brackets "{ }" are used for this dynamic part. The final "\n", instead, is used to leave a blank line after the printed text.

```python
diff_words = len(set(hamlet))
print(f"Hamlet contains {diff_words} different words\n")
```

Similar to the previous: The only difference is that we now convert the list into a set, using the <em>set</em> function, before computing its length. Since sets cannot have duplicates, the system is removing them during the conversion. This is why the number is much smaller.



```python
diff_words_non_case = len(set(w.lower() for w in hamlet))
print(f"Hamlet contains {diff_words_non_case} different words, if we make it non-case-sensitive\n")
```

This is a bit trickier :D 

Well, the concept is easy: in the code before, we considered words such as "However" and "however" to be different, as they are technically different strings. In other words, our duplicates removal was <em>case-sensitive</em>. Now, we want to make it <em>non-case-sensitive</em> (or case-insensitive?).

The code is the tricky part, in particular the one inside the <em>set()</em>, which can be read as follows:<br> <em>For each word in the "hamlet" list, temporarily called "w", apply the lower() <strong>method</strong> to it, in order to make it lower case</em>.<br>
In this way, both "However" and "however" will be turned into "however", and then be considered as duplicates when the set() function is applied.

<em>Note</em>: Here, <em>w</em> is used as a working variable, which contains all the elements one after the other. This allows to repeat the same operation on all the elements one by one.

<em>Other note</em>: A <strong>method</strong> is very similar to a <strong>function</strong>, but it is applied on an object, using the dot ".", as in <em>w.lower()</em>.

```python
hamlet_sentences = gutenberg.sents("shakespeare-hamlet.txt")

sentences = len(hamlet_sentences)
print(f"Hamlet contains {sentences} sentences\n")
```

Same as before, but now we get all the sentences of the Hamlet, not the words.

```python
first = hamlet_sentences[0]
print(f"The first sentence in the Hamlet (as stored in Python) is:\n{first}\n")
```

When you have a list, you can access single elements by their position (more formally, <em>index</em>). Note that indices start from 0, so <em>hamlet_sentences[0]</em> is the first sentence (while, for example, <em>hamlet_sentences[8]</em> is the ninth -- you may try changing the index in the box code, run it, and see the result).

You may have noticed that the sentence is not stored as a single block of text, but rather as a list of words (technically, <em>hamlet_sentences</em> is a list of lists). In the next code we print the corresponding block of text, using the <em>join</em> method. 

```python
first = " ".join(hamlet_sentences[0])
print(f"The first sentence in the Hamlet (as a nicer string) is:\n{first}\n")
```

It may look scary, but it is not :) We are basically asking to join all the elements of the list composing the sentence into a single string of text by separating the elements with a blank space " ".

We are almost done, but I would like to add a small piece of code to make something more advanced, to show a small hint of the power of the library.

The next two instructions compute a frequency distribution of the words in the Hamlet, with the <em>FreqDist()</em> function, then print the 20 most common words with their number of occurrences, and finally plot the cumulative distribution ("," appears 2892 times -- "," and "." appear 2892+1886=4778 times in total -- ... -- the first 20 words appear more than 13000 times in total, that is, they are more than one third of the total 37360 words in the text). Of course this does not say much, being mostly punctuation, conjunctions, or prepositions. We would need to first clean the text by filtering out these common words, but not today :D

```python
fdist = FreqDist(hamlet)

print("These are the 20 most common terms in the Hamlet")
print(fdist.most_common(20))
```

```python
fdist.plot(20, cumulative = True)
```
